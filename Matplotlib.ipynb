```python
numpydizisi = np.linspace(0,10,20)
numpydizisi
```

```python
numpydizisi2 = numpydizisi ** 3
numpydizisi2
```

```python
plt.plot(numpydizisi, numpydizisi2, "b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizisi, numpydizisi2, "r*-")
plt.subplot(1,2,2)
plt.plot(numpydizisi2, numpydizisi, "g--")
```

```python
benimfigure = plt.figure()
figureAxes = benimfigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(numpydizisi, numpydizisi2, "g")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni bveri ismi")
figureAxes.set_title("grafik basligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.1,0.5,0.2,0.2])

eksen1.plot(numpydizisi, numpydizisi2, "g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik basligi")

eksen2.plot(numpydizisi2, numpydizisi, "b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("kucuk grafik basligi")
```

```python
(benimfigure, benimeksen) = plt.subplots(nrows=1, ncols=2)
type(benimeksen)

for eksen in benimeksen:
    eksen.plot(numpydizisi, numpydizisi2, "g")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()
```

```python
# dpi 150
newfigure = plt.figure(figsize=(6,6), dpi=100)
newfigure = newfigure.add_axes([0.1,0.1,0.9,0.9])

newfigure.plot(numpydizisi, numpydizisi ** 2, label= "numpydizisi ** 2")
newfigure.plot(numpydizisi, numpydizisi ** 3, label= "numpydizisi ** 3" )
newfigure.legend(loc=1)
```
