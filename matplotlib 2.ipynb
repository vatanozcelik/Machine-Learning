## gorsellik

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpydizisi = np.linspace(0,10,20)
```

```python
numpydizisi
```

```python
numpydizisi2 = numpydizisi ** 2
```

```python
(benimfigire, benimeksen) = plt.subplots()
benimeksen.plot(numpydizisi, numpydizisi2, color = "#FF0000",alpha=1)
benimeksen.plot(numpydizisi2, numpydizisi, color = "#000000")
```

```python
(newfigure, neweksen) = plt.subplots()
# linewidth kalinlik
neweksen.plot(numpydizisi, numpydizisi + 2, color="blue", linewidth=8.0)
neweksen.plot(numpydizisi, numpydizisi + 3, color="yellow", linewidth=3.0)

neweksen.plot(numpydizisi, numpydizisi + 8, color="green", linewidth=12.0, linestyle="-.")
neweksen.plot(numpydizisi, numpydizisi + 12, color="black", linewidth=22.0, linestyle=":",markerfacecolor="red", marker="+", markersize=12.0)
neweksen.plot(numpydizisi, numpydizisi + 5, color="grey", linestyle="--")
```
