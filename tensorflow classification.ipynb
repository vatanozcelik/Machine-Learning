```python
y = data["Type"].values
x = data.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=1, activation = "sigmoid"))
# sigmoid output 1 or 0

model.compile(loss="binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=700, validation_data=(X_test, y_test), verbose=1)
```

```python
model.history.history
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
model=Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=15, activation = "relu"))
model.add(Dense(units=1, activation = "sigmoid"))
# sigmoid output 1 or 0

model.compile(loss="binary_crossentropy", optimizer = "adam")
```

## early stopping

```python
earlystopping = EarlyStopping(monitor = "val_loss", mode="min",verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=700, validation_data=(X_test, y_test), verbose=1, callbacks=[earlystopping])
```

```python
modelloss = pd.DataFrame(model.history.history)
```

```python
modelloss.plot()
```

### dropout

```python
model=Sequential()

model.add(Dense(units=30, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1, activation = "sigmoid"))
# sigmoid output 1 or 0


model.compile(loss="binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=700, validation_data=(X_test, y_test), verbose=1, callbacks=[earlystopping])
```

```python
modellosss = pd.DataFrame(model.history.history)
```

```python
modellosss.plot()
```

```python
prediction = model.predict_classes(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prediction))
```

```python
print(confusion_matrix(y_test, prediction))
```
